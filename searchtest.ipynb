```python
class Equities:
  Equities = fd.Equities()  # Assuming fd.Equities() is a valid object that can perform searches

  @staticmethod
  def search(country, industry, market_cap, currency): # Add exchange as an argument with a default value
    country = country.capitalize()
    industry = industry.capitalize()
    market_cap = market_cap.capitalize()
    currency = currency.upper()

    # Assuming Equities.search() is the actual method to perform the search
    # Pass the exchange argument, or None if not provided
    market = Equities.Equities.search(country=country, industry=industry, market_cap=market_cap ,currency=currency)

    if not market.empty:  # Check if the result DataFrame is empty
      return market
    else:
      print("No results found")
      return None

# Example usage
country = "United States"
industry = "Semiconductors & Semiconductor Equipment"
market_cap = "Mega Cap"
currency = "USD"

market = Equities.search(country, industry, market_cap, currency) # Pass the exchange value to the search function
market
```

```python
class ETFs:
  ETFs = fd.ETFs()  # Assuming fd.Equities() is a valid object that can perform searches

  @staticmethod
  def search(category, market, currency):
    category = category.capitalize()
    market = market.capitalize()
    currency = currency.upper()

    # Assuming Equities.search() is the actual method to perform the search
    market = ETFs.ETFs.search(category=category, market=market, currency=currency)

    if not market.empty:  # Check if the result DataFrame is empty
      return market
    else:
      print("No results found")
      return None

# Example usage
category = "Technology"
market = "us_market"
currency = "USD"

market = ETFs.search(category, market, currency)
market
```

```python
class Funds:
  Funds = fd.Funds()  # Assuming fd.Equities() is a valid object that can perform searches

  @staticmethod
  def search(category, market, currency):
    category = category.capitalize()
    market = market.capitalize()
    currency = currency.upper()

    # Assuming Equities.search() is the actual method to perform the search
    market = Funds.Funds.search(category=category, market=market, currency=currency)

    if not market.empty:  # Check if the result DataFrame is empty
      return market
    else:
      print("No results found")
      return None

# Example usage
category = "Bonds"
market = "us_market"
currency = "USD"

market = Funds.search(category, market, currency)
market
```
